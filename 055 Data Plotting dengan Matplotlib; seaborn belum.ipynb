# Bagian 5: Data Plotting dengan Matplotlib

matplotlib adalah paket yang dapat digunakan untuk membuat grafik 2d, seperti histogram, scatter plot, dan grafik-grafik lainnya pada python. matplotlib dikembangkan sedemikian rupa agar memiliki sintaks yang mudah dipahami, dan memiliki kimiripan dengan sintax pada MatLab.

```python
import matplotlib.pyplot as plt
```

### Mengatur Skal pada Axis dan membalikkannya

Skala pada axis dapat diubah dari linear menjadi Logaritmik dengan fungsi: plt.gca().set_xscale('log') atau plt.gca().set_yscale('log')

Untuk membalikkan axis dapat menggunakan plt.gca().invert_xaxis() atau plt.gca().invert_yaxis()

```python
plt.hist(y)
plt.ylim(0.2,2000)
plt.xticks([0,2,2.5,4], 
           ['satu', 'dua', 'tiga', 'empat'], 
           rotation=90);
```

```python
plt.hist(y)
plt.ylim(0.2,2000)
plt.xticks([0,2,2.5,4], 
           ['satu', 'dua', 'tiga', 'empat'], 
           rotation=90);

plt.gca().set_yscale('log')
plt.gca().invert_yaxis()
```

## Lineplot

Lineplot adalah diagram garis. Data yang diperlukan adalah data X dan Y dalam bentuk array atau dalam bentuk DataFrame. Note: X dan Y harus memiliki panjang yang sama.
Perintah yang digunakan untuk membuat lineplot adalah:

plt.plot(args)
X: data dimensi X
y: data dimensi y
format garis dalam bentuk teks. Contoh format garis:
'-g' = garis tegas berwarna hijau
'--r' = garis putus-putus berwarna merah

```python
plt.plot(X,y)
```

```python
"creating dummy data"
X1 = [1,3,2,9]
y1 = [9,7,4,1]

plt.plot(X1,y1,'--g')
```

```python
plt.plot(X1,y1,'--r')
```

```python
"creating dummy data"
i = [1,4,8,10]
s = [8,12,16,20]

plt.plot(i,s,'--g')
```

```python
plt.plot(i,s,linewidth=3, marker = 'o', markerfacecolor='r')
```

```python
plt.plot(X1,y1,linewidth=0.5, marker ='o', markerfacecolor='r')
```

## Scatter Plot

Scatter plot adalah diagram titik. Data yang diperlukan adalah data X dan Y dalam bentuk array atau dalam bentuk DataFrame. Note: X dan Y harus memiliki panjang yang sama.
Perintah yang digunakan untuk membuat lineplot adalah:

- plt.scatter(args)
- X: data dimensi X
- y: data dimensi y
- c: warna (bisa dalam bentuk array berukuran yang sama dengan X)
- s: ukuran marker (bisa dalam bentuk array berukuran yang sama dengan X)
- alpha: transparansi dari data (bisa dalam bentuk array berukuran yang sama     dengan X)

```python
plt.scatter(i,s,c='b')
```

```python
'''inisasi data random'''
import numpy as np
X = np.random.random(size=(10000,1))
y = np.random.normal(size=(10000,1))
```

```python
plt.scatter(X,y,c='y')
```

```python
plt.figure(figsize=(15,10))
plt.scatter(X,y,s=X*50,alpha=0.1,c='b')
```

```python
plt.figure(figsize=(15,10))
plt.scatter(X,y,s=X*20,alpha=0.6,c='b')
```

## 4 Image plot

Image plot adalah grafik benbentuk grid dengan ukuran (i * j). Masing-masing nilai pada sel diwakili dengan warna. Untuk membuat grafik diperlukan Array dengan ukuran (i * j). Perintah yang digunakan adalah:

- plt.imshow(args)
argumen yang dapat digunakan:
- data: data (dalam bentuk i * j)
- cmap: colormap
- interpolation: interpolasi gambar

```python
X1 = np.array([[1,6,7,9,12,14,-2,2,5,5,7,9]])
plt.imshow(X1)
```

```python
X1 = X1.reshape(-1,3)
plt.imshow(X1)
```

```python
X1 = X1.reshape(-1,2)
plt.imshow(X1)
```

### Menggunakan Subplot

Subplot adalah perintah untuk membagi figure kedalam panel dengan ukuran (i * j). dengan sintax:

plt.subplot(i,j,n)
i = banyak kolom
j = banyak baris
n = urutan

```python
plt.figure(figsize=(int,int))

plt.subplot(i,j,n)
plt.graphic_command1()

plt.subplot(i,j,n+1)
plt.graphic_command2()

plt.subplot(i,j,n+2)
plt.graphic_command3()

plt.subplot(i,j,n+3)
plt.graphic_command4()
.... 
```

```python
'''creating dummy data'''
X1 = [1,6,7,9,12,14]
y1 = [-2,2,5,5,7,9]

X2 = [2,1,3,4,7]
y2 = [9,7,3,2,-1]
```

```python
plt.figure(figsize=(20,5))

plt.subplot(1,2,1)
plt.scatter(X1,y1)
plt.title('grafik 1')

plt.subplot(1,2,2)
plt.plot(X1,y1,'r')
plt.title('grafik 2')
```
